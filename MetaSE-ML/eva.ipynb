# baseline - svm-linear

```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## 1.1 Heart model version 2

```python
GEM_name = "ML+GEM_Lung_version1"

# load result
result = pd.read_csv(f'out/{GEM_name}_svmLinear_out.csv')

# 100 iteration, se-pos>10
result = result[result.n_sample > 11]
result.mean()
```

```python
fig = go.Figure(go.Histogram(x=result.auroc_test, nbinsx=40, name='testing'))
fig.add_trace(go.Histogram(x=result.auroc_train, nbinsx=40, name='training'))
fig.update_layout(xaxis_title='auroc', yaxis_title='count of side effect per auroc bin')
fig.show()
```

```python
px.scatter(result, x='n_sample', y=['auroc_test', 'auroc_train'], opacity=0.7)
```

```python
result[result.auroc_test > 0.7].sort_values(['auroc_test', 'se_name'], ascending=False)
```
